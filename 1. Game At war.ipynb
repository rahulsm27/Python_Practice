```python
# Programme to crate war game using pythoon OOPS

import random

suits = ['Diamond','Hearts','Clubs', 'Spades']
rank = ['Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace']
value = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':11,'Queen':12,'King':13,'Ace':14}


class card():
    
    def __init__(self,suits,rank):
        self.rank = rank
        self.suits = suits
        self.value = value[rank]
        
    def __str__(self):
        return (f"I am {self.suits} of {self.rank}")
    

class deck():
    
#     deckno = []
    def __init__(self):
         self.deck_list = []
    
    def create_deck(self):
        for i in suits:
            for j in rank:
                card_created = card(i,j)
                self.deck_list.append(card_created)
               # print (deck_list)
        return self.deck_list
    
    def shuffle(self):
        random.shuffle(self.deck_list)
        return self.deck_list
    
    
class player():
    
    def __init__(self, name):
        self.player_name = name
        
    def __str__(self):
        return (f"I am {self.player_name}")
        
```
